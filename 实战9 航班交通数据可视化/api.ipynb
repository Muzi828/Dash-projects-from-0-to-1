```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

data=pd.read_csv('data/2011_february_us_airport_traffic.csv')
data.head()
```

```python
#获取州标签的唯一值，并核实无缺失值None
state_ls = data['state'].unique()
print(state_ls)
```

```python
data['country'].unique()
```

```python
#获取州和机场的交通量
state_airport_data = data.groupby(['state','airport','lat','long'])['cnt'].sum().reset_index()
state_airport_data 
```

```python
#获取最多的机场所在的州
max_airport= state_airport_data ['state'].value_counts().index[0]
max_airport
```
